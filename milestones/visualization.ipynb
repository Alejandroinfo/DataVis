```python

df = df.copy()
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.loc[:, 'date'] = pd.to_datetime(df['date'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')
df.loc[:, 'year'] = df['date'].dt.year
df.loc[:, 'month'] = df['date'].dt.month
df.loc[:, 'hour'] = df['date'].dt.hour
df.loc[:, 'day'] = df['date'].dt.day
ddf = df.dropna(subset=['latitude', 'longitude']).copy()

# --- Boolean conversions ---
df.loc[:, 'arrest'] = df['arrest'].astype(bool)
df.loc[:, 'domestic'] = df['domestic'].astype(bool)

# --- Optional sampling for speed ---
df_sample = df.sample(min(10000, len(df)), random_state=42).copy()
alt.themes.enable('dark')
```

```python
crime_type_chart = (
    alt.Chart(df_sample)
    .mark_bar()
    .encode(
        x=alt.X('count():Q', title='Number of Crimes'),
        y=alt.Y('primary_type:N', sort='-x', title='Crime Type'),
        color=alt.Color('primary_type:N', legend=None),
        tooltip=['primary_type', 'count()']
    )
    .properties(
        title='Most Common Crime Types in Chicago (2024)',
        width=600, height=400
    )
)
crime_type_chart
```
